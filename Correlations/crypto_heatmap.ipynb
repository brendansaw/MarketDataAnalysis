```python
import pandas as pd
import re
import seaborn
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import os
import correlations
```

```python
file_dir = os.path.dirname(os.path.realpath('__file__'))
all_data_files = [file for file in os.listdir(Path(os.path.join(file_dir, "../data/crypto")))]
all_data_files
```

```python
daily_pickle_files = []
for file in all_data_files:
    if len(file) > 6 and file[-6:] == "_D.pkl":
        daily_pickle_files.append(file)
daily_pickle_files
```

```python
dfs = []
for file in daily_pickle_files:
    df = correlations.get_crypto_pickle_df(f"../data/crypto/{file}").rename(correlations.filepath_to_filename(file))
    dfs.append(df)
dfs
```

```python
joined = pd.concat(dfs, axis=1).dropna()
joined
```

```python
correlation_pairs = joined.corr()
correlation_pairs
```

```python
hm = seaborn.heatmap(correlation_pairs)
# plt.show()
plt.title("Crypto Pair Co-integration")
plt.ylabel("Pair 1")
plt.xlabel("Pair 2")
# bottom, top = hm.get_ylim()
# hm.set_ylim(bottom + 0.5, top - 0.5)
plt.savefig(fname="crypto_heatmap.png", format="png", bbox_inches='tight')
```
